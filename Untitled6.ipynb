```python
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
df = pd.read_csv('BBC_list_backup.csv')
df.shape
```

```python
bechdel = pd.read_csv('mylist_backup.csv')
bechdel.shape
```

```python
df_join = pd.merge(df, bechdel, how='left', on="movie")
df_join.head()
```

```python
df_join.shape
df_join = df_join.drop_duplicates()
len(df_join)
```

```python
df_gender = pd.read_csv('Male_directors.backup.csv')
df_gender.head()
```

```python
df_merge = pd.merge(df_join, df_gender, how='left', on="director")
```

```python
df_merge.head(20)
```

```python
#backup your dataframes
df_merge.to_csv('Final_dataframe.csv', index=False)
```
