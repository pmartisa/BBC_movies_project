```python
from bs4 import BeautifulSoup
import requests
import re
import numpy as np
import pandas as pd
from time import sleep
from random import randint
```

```python
my_url = "https://www.bbc.com/culture/article/20191125-the-100-greatest-films-directed-by-women-poll"
response = requests.get(my_url).content
```

```python
soup_doc = BeautifulSoup(response, "html.parser")
print(soup_doc.prettify())
```

```python
all_info = soup_doc.find(class_="body-text-card__text body-text-card__text--culture body-text-card__text--drop-capped body-text-card__text--flush-text")
movie_info = all_info.find_all(string=True)
female_directors = []
for movie in movie_info[35:135]:
    movie_director = re.findall(r"[()](\w.*)[,]", movie)
   # director_names = [movie_director[0]]
    gender = "F"
    values = [movie_director[0], gender]
    female_directors.append(values)
```
