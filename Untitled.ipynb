```python
# You can just use BeautifulSoup now
doc = BeautifulSoup(driver.page_source)

for title in doc.select(".title"):
    print(title.text)
```

```python
# Sadly can't do this any more
# driver.find_elements_by_class_name("title")
# By.CLASS_NAME or By.ID or By.CSS_SELECTOR
titles = driver.find_elements(By.CLASS_NAME, "title")
for title in titles:
    print(title.text)
```

```python
# Go find class='title'
# then find the ul inside of that
# then find all of the li's inside of that
books = doc.select(".title-list ul li")
len(books)
```

```python
dataset = []

for book in books:
    data = {}
    data['title'] = book.select_one(".title p a").text
    data['names'] = book.select_one("span.name").text
    data['score'] = book.select_one(".score").text
    # Every time through the loop
    # we add another dictionary of a book
    # to our dataset
    dataset.append(data)

df = pd.DataFrame(dataset)
df
```
